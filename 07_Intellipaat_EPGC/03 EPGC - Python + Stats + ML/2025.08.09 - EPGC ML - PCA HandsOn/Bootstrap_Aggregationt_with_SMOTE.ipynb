# SMOTE (Synthetic Minority Oversampling Technique)
* SMOTE creates new data points for the minority class by combining nearby existing ones.
* It helps balance the dataset, making the model better at identifying the minority class, which improves recall.

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE

smote = SMOTE(random_state=0)

X_resample,y_resample = smote.fit_resample(X_train, y_train)

# Train your model with the resampled data
classifier1.fit(X_resample,y_resample)   #classifier1.fit(X_train,y_train)

# Evaluate the model
y_pred_smote = classifier1.predict(X_test)
```
