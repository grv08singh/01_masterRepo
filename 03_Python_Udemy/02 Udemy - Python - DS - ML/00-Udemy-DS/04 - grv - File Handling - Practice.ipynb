```python
# File opening modes
# r : read mode. gives error if file is not present.
# a : append mode. appends data at the end. creates file if not present.
# w : write mode. over-writes the existing data. creates file if not present.
# x : create mode. creates a new file. returns an error if file already exists.
```

```python
f = open("testFile.txt",'r')
print(f.read())
```

```python
#print first 4 characters
f = open("testFile.txt",'r')
print(f.read(4))
```

```python
#print first line
f = open("testFile.txt",'r')
print(f.readline())
```

```python
#print second line
print(f.readline())
```

```python
#read line by line using for loop
f = open("testFile.txt",'r')
for x in f:
    print(x)
```

```python
f = open("testFile.txt",'a')
f.write("\nLine appended.")
```
