#### **04-DataFrames - Part 3**

```python
#Index Levels
outside = ['G1','G1','G1','G2','G2','G2']
```

```python
inside = [1,2,3,1,2,3]
```

```python
hier_index = list(zip(outside, inside))
```

```python
hier_index
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
df = pd.DataFrame(randn(6,2), hier_index, ['A', 'B'])
```

```python
df
```

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```

```python
df
```

```python
df.index.names = ['Groups', 'Num']
```

```python
df
```

```python
df.loc['G2']
```

```python
df.loc['G2'].loc[2]
```

```python
df.loc['G2'].loc[2]['B']
```
