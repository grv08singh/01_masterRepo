# Data Generation

```python
a = np.array([0,0,0,0,0])
print(a)
```

```python
# zeros 
# 5 - rows, 6 - column
# by default the data will be in form of float
a = np.zeros((5,6))
print(a)
print(type(a))
```

```python
# ones 
# 5 - rows, 6 - column
# by default the data will be in form of float
a = np.ones((5,6))
print(a)
print(type(a))
```

```python
# twos 
# 5 - rows, 6 - column
# by default the data will be in form of float
a = np.ones((5,6))* 2
print(a)
print(type(a))
```

```python
# 5 - rows, 6 - column
# by default the data will be in form of float
a = np.ones((5,6)) + 54
print(a)
print(type(a))
```

```python
# full 

# 5 - rows, 6 - column, 7 - number
# by default the data will be in form of float
a = np.full((5,6), 7)
print(a)
print(type(a))
```

```python
# full 

# 5 - rows, 6 - column, 7.3 - number
# by default the data will be in form of float
a = np.full((5,6), 7.3)
print(a)
print(type(a))
```

```python
# linspace - equal spacing 
# 0 - start, 20 - end point , 11 - no of point 
# start and end point are included

a = np.linspace(0,20,10)
print(a)
```

```python
a = np.linspace(0,20,25)
print(a)
```

```python
# arange 
# 0 - start, 20 - end , 3 - step size 
# end is not included 

a = np.arange(0,20,3)
print(a)
```

```python
a = np.arange(0,20,5)
print(a)
```

```python
# random 
# values will always between 0 and 1 
# 1 - random - module 2- random - function
# 40 - no. of points

a = np.random.random(40)
print(a)
```

```python
# randint
# 0 - start, 40 - end 
a = np.random.randint(0,40)
print(a)
```

# Array Inspection

```python
a = np.array(20)
print(a)
print(type(a))
print(np.size(a)) # count of elements present in the array
print(np.ndim(a)) # dimension of the array
print(np.shape(a)) # will give you the shape of the array
```

```python
a = np.array([20,10,30,40,50])
print(a)
print(type(a))
print(np.size(a)) # count of elements present in the array
print(np.ndim(a)) # dimension of the array
print(np.shape(a)) # will give you the shape of the array
```

```python
a = np.array([[20,10,30],
              [40,50,60]])
print(a)
print(type(a))
print(np.size(a)) # count of elements present in the array
print(np.ndim(a)) # dimension of the array
print(np.shape(a)) # will give you the shape of the array
```

```python
a = np.array([[20,10,30,50],
              [40,50,60,70],
              [90,80,70,90]])
print(a)
print(type(a))
print(np.size(a)) # count of elements present in the array
print(np.ndim(a)) # dimension of the array
print(np.shape(a)) # will give you the shape of the array
```

```python
a = np.array([[[20,10,30,50],
              [40,50,60,70],
              [90,80,70,90]]])
print(a)
print(type(a))
print(np.size(a)) # count of elements present in the array
print(np.ndim(a)) # dimension of the array
print(np.shape(a)) # will give you the shape of the array
```

```python
a = np.array([
              [[20,10,30,50],[40,50,60,70],[90,80,70,90]],
              [[20,10,30,50],[40,50,60,70],[90,80,70,90]]
             ])
print(a)
print(type(a))
print(np.size(a)) # count of elements present in the array
print(np.ndim(a)) # dimension of the array
print(np.shape(a)) # will give you the shape of the array
```

# Comparision of Array

```python
# element wise comparision

a = [1,2,3]
b = [1,3,4]
np.equal(a,b)
```

```python
a = [1,2,3]
b = [1,2,3]
np.equal(a,b)
```

```python
# array wise comparision

a = [1,2,3]
b = [1,3,4]
np.array_equal(a,b)
```

```python
a = [1,2,3]
b = [1,2,3]
np.array_equal(a,b)
```
