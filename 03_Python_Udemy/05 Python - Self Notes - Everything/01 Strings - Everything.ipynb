## Printing Strings

```python
name = 'Adam'
```

```python
age = 24
```

```python
work = 9.85
```

#### using multiple formatting

```python
print('%d year old %s has completed %3.2f percent of work.'%(age, name, work))
```

#### using .format()

```python
print('{age} year old {name} has completed {work} percent of work.'.format(age=31, name='Jon', work=5.73))
```

#### using f-strings - best way

```python
print(f'{age} year old {name} has completed {work} percent of work.')
```
