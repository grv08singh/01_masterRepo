# 2025.05.10 - Data Visualization (Hands-on 1)¶

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'cars_2.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.describe()
```

```python
df.info()
```

```python
df['cubicinches'].unique()
```

```python
df['cubicinches'] = df['cubicinches'].replace(' ', np.nan)
```

```python
df['cubicinches'].unique()
```

```python
df.info()
```

```python
df['cubicinches'] = pd.to_numeric(df['cubicinches'], errors = 'coerce')
```

```python
df.info()
```

```python
df['weightlbs'].unique()
```

```python
df['weightlbs'] = pd.to_numeric(df['weightlbs'].replace(' ', np.nan), errors = 'coerce')
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns
```

```python
# 1) Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg,
# hp-time-to-60) with respect to each brand.
```

```python
fig, axis = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
#using sns
sns.scatterplot(data=df, x='mpg', y='hp', hue='brand', ax = axis[0,0])
sns.scatterplot(data=df, x='mpg', y='weightlbs', hue='brand', ax = axis[0,1])
sns.scatterplot(data=df, x='hp', y='mpg', hue='brand', ax = axis[1,0])
sns.scatterplot(data=df, x='hp', y='time-to-60', hue='brand', ax = axis[1,1])

plt.show()
```

```python
plt.figure(figsize=(12,8))
#using plt
plt.subplot(2,2,1).scatter(x=df['mpg'], y=df['hp'])
plt.xlabel('mpg')
plt.ylabel('hp')
plt.subplot(2,2,2).scatter(x=df['mpg'], y=df['weightlbs'])
plt.xlabel('mpg')
plt.ylabel('weightlbs')
plt.subplot(2,2,3).scatter(x=df['hp'], y=df['mpg'])
plt.xlabel('hp')
plt.ylabel('mpg')
plt.subplot(2,2,4).scatter(x=df['hp'], y=df['time-to-60'])
plt.xlabel('hp')
plt.ylabel('time-to-60')

plt.show()
```

```python
df.columns
```

```python
#2.a) Create a bar plot that shows the visual representation of hp, mpg, weight,
#time-to-60 with respect to the number of cylinders for each of the brands.
```

```python
fig, axis = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
#using sns
sns.barplot(data=df, x='cylinders', y='hp', hue='brand', ax = axis[0,0])
sns.barplot(data=df, x='cylinders', y='mpg', hue='brand', ax = axis[0,1])
sns.barplot(data=df, x='cylinders', y='weightlbs', hue='brand', ax = axis[1,0])
sns.barplot(data=df, x='cylinders', y='time-to-60', hue='brand', ax = axis[1,1])

plt.show()
```

```python
plt.figure(figsize=(12,8))

#using plt
plt.subplot(2,2,1).bar(x=df['cylinders'], height=df['hp'])
plt.xlabel('cylinders')
plt.ylabel('hp')
plt.subplot(2,2,2).bar(x=df['cylinders'], height=df['mpg'])
plt.xlabel('cylinders')
plt.ylabel('mpg')
plt.subplot(2,2,3).bar(x=df['cylinders'], height=df['weightlbs'])
plt.xlabel('cylinders')
plt.ylabel('weightlbs')
plt.subplot(2,2,4).bar(x=df['cylinders'], height=df['time-to-60'])
plt.xlabel('cylinders')
plt.ylabel('time-to-60')

plt.show()
```

```python
# 2.b) Create a bar plot that shows the visual representation of hp, mpg, weight and
# time-to-60 with respect to the years, for each brand
```

```python
fig, axis = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
#using sns
sns.barplot(data=df, x='year', y='hp', hue='brand', ax = axis[0,0])
sns.barplot(data=df, x='year', y='mpg', hue='brand', ax = axis[0,1])
sns.barplot(data=df, x='year', y='weightlbs', hue='brand', ax = axis[1,0])
sns.barplot(data=df, x='year', y='time-to-60', hue='brand', ax = axis[1,1])

plt.show()
```

```python
plt.figure(figsize=(12,8))

#using plt
plt.subplot(2,2,1).bar(x=df['year'], height=df['hp'])
plt.xlabel('year')
plt.ylabel('hp')
plt.subplot(2,2,2).bar(x=df['year'], height=df['mpg'])
plt.xlabel('year')
plt.ylabel('mpg')
plt.subplot(2,2,3).bar(x=df['year'], height=df['weightlbs'])
plt.xlabel('year')
plt.ylabel('weightlbs')
plt.subplot(2,2,4).bar(x=df['year'], height=df['time-to-60'])
plt.xlabel('year')
plt.ylabel('time-to-60')

plt.show()
```

```python
# 6) Create histograms for hp, mpg, time-to-60, weight lbs, cubic inches, year, with
# respect to brand
```

```python
fig, axis = plt.subplots(nrows=2, ncols=3, figsize=(12,8))
#using sns
sns.histplot(data=df, x='hp', hue='brand', ax = axis[0,0])
sns.histplot(data=df, x='mpg', hue='brand', ax = axis[0,1])
sns.histplot(data=df, x='time-to-60', hue='brand', ax = axis[0,2])
sns.histplot(data=df, x='weightlbs', hue='brand', ax = axis[1,0])
sns.histplot(data=df, x='cubicinches', hue='brand', ax = axis[1,1])
sns.histplot(data=df, x='year', hue='brand', ax = axis[1,2])

plt.show()
```

```python
plt.figure(figsize=(12,8))

#using plt
plt.subplot(2,3,1).hist(x=df['hp'], bins=8, edgecolor="white")
plt.xlabel('hp')
plt.ylabel('frequency')
plt.subplot(2,3,2).hist(x=df['mpg'], bins=8, edgecolor="white")
plt.xlabel('mpg')
plt.ylabel('frequency')
plt.subplot(2,3,3).hist(x=df['time-to-60'], bins=8, edgecolor="white")
plt.xlabel('time-to-60')
plt.ylabel('frequency')
plt.subplot(2,3,4).hist(x=df['weightlbs'], bins=8, edgecolor="white")
plt.xlabel('weightlbs')
plt.ylabel('frequency')
plt.subplot(2,3,5).hist(x=df['cubicinches'], bins=8, edgecolor="white")
plt.xlabel('cubicinches')
plt.ylabel('frequency')
plt.subplot(2,3,6).hist(x=df['year'], bins=8, edgecolor="white")
plt.xlabel('year')
plt.ylabel('frequency')

plt.show()
```

```python
# 7) Line plot between hp-mpg, hp-(time-to-60), mpg-(time-to-60), hp-weight lbs,
# mpg-weight lbs, cubic inches-weight lbs
```

```python
fig, axis = plt.subplots(nrows=2, ncols=3, figsize=(12,6))
#using sns
sns.lineplot(data=df, x='hp', y='mpg', ax = axis[0,0])
sns.lineplot(data=df, x='hp', y='time-to-60', ax = axis[0,1])
sns.lineplot(data=df, x='mpg', y='time-to-60', ax = axis[0,2])
sns.lineplot(data=df, x='hp', y='weightlbs', ax = axis[1,0])
sns.lineplot(data=df, x='mpg', y='weightlbs', ax = axis[1,1])
sns.lineplot(data=df, x='cubicinches', y='weightlbs', ax = axis[1,2])

plt.show()
```

```python
plt.figure(figsize=(12,6))
#using plt
plt.plot()
plt.subplot(2,3,1).plot(df['hp'], df['mpg'])
plt.xlabel('hp')
plt.ylabel('mpg')
plt.subplot(2,3,2).plot(df['hp'], df['time-to-60'])
plt.xlabel('hp')
plt.ylabel('time-to-60')
plt.subplot(2,3,3).plot(df['mpg'], df['time-to-60'])
plt.xlabel('mpg')
plt.ylabel('time-to-60')
plt.subplot(2,3,4).plot(df['hp'], df['weightlbs'])
plt.xlabel('hp')
plt.ylabel('weightlbs')
plt.subplot(2,3,5).plot(df['mpg'], df['weightlbs'])
plt.xlabel('mpg')
plt.ylabel('weightlbs')
plt.subplot(2,3,6).plot(df['cubicinches'], df['weightlbs'])
plt.xlabel('cubicinches')
plt.ylabel('weightlbs')

plt.show()
```
