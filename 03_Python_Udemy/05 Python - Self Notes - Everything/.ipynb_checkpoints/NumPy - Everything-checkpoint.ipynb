## Create an array of n equally spaced points between start and end both included

### np.linspace(start, end [, n]): n number of equally spaced points between start and end

```python
# linspace - equal spacing 
# 0 - start, 20 - end point , 11 - no of point 
# start and end point are included

a = np.linspace(0,20,10)
print(a)
```

```python
a = np.linspace(0,20,25)
print(a)
```

## Create a list of numbers with step_size as difference between two consecutive numbers.

### np.arange([start,] end_point + 1 [, step_size]): 

```python
# arange 
# 0 - start, 20 - end , 3 - step size 
# end is not included 

a = np.arange(0,20,3)
print(a)
```

```python
a = np.arange(0,20,5)
print(a)
```

```python
np.arange(10,51,2) # even numbers
```

```python
np.arange(11,51,2) # odd numbers
```

## np.diag(arr): Diagonal of array

```python
lst = [[7,3,9],[1,0,2],[4,8,6]]
```

```python
arr = np.array(lst)
```

```python
arr
```

## one of the diagonals

```python
arr_diag1 = np.diag(arr)
```

```python
arr_diag1
```

## other diagonal

### solution 1: first flip left to right

```python
np.fliplr(arr)
```

```python
arr_diag2 = np.diag(np.fliplr(arr))
```

```python
arr_diag2
```

### solution 2: first rotate the array 90 degrees : anticlock-wise

```python
np.rot90(arr)
```

```python
arr_diag2 = np.diag(np.rot90(arr))
```

```python
arr_diag2
```

## Array Inspection

```python
a = np.array([[20,10,30],
              [40,50,60]])
print(a)
print()
print(type(a))
print()
print(np.size(a)) # count of elements present in the array
print()
print(np.ndim(a)) # dimension of the array
print()
print(np.shape(a)) # will give you the shape of the array
```

```python
a = np.array([[20,10,30,50],
              [40,50,60,70],
              [90,80,70,90]])
print(a)
print(type(a))
print(np.size(a)) # count of elements present in the array
print(np.ndim(a)) # dimension of the array
print(np.shape(a)) # will give you the shape of the array
```

## Slicing

```python
# slicing
a = np.array([10,20,30,40,50])
# [10,20]
# a[start:end+1:step]
a[0:2:1]
```

```python
a = np.array([10,20,30,40,50])
# [30,40]
a[2:4:1]
```

```python
a = np.array([10,20,30,40,50])
# [20,40]
a[1:4:2]
```

```python
# 2d
# indexing
a = np.array([[10,20,30],
              [40,50,60]])

# a[rows,column]
# 10
a[0,0]
```

```python
# 50
a[1,1]
```

```python
# 30
a[0,2]
```

```python
# 3d
# indexing
a = np.array([[[10,20,30],
              [40,50,60],
              [90,80,70]]])

# a[array,rows,column]
# 10 
a[0,0,0]
```

```python
a = np.array([[[10,20,30],
              [40,50,60],
              [90,80,70]]])
# 60
a[0,1,2]
```

```python
# 70
a[0,2,2]
```

```python
# 2d
# slicing
a = np.array([[10,20,30],
              [40,50,60]])

#[[10,20]]

# a[rstart:rend+1:rstep, cstart:cend+1:cstep]
a[0:1:1,0:2:1]
```

```python
a = np.array([[10,20,30],
              [40,50,60]])

# [[50,60]]
a[1:2:1,1:3:1]
```

```python
# 3d
a = np.array([[[10,20,30],
              [40,50,60]]])
# [[[50,60]]]

a[0:1:1, 1:2:1,1:3:1]
```
