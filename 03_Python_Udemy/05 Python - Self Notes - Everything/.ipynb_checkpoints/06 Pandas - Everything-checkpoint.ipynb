## Pandas Series

```python
# Creating list
labels = ['a','b','c']
my_list = [10,20,30]

#Creating Array
arr = np.array([10,20,30])

#Creating Dictionary
dict = {'a':10,'b':20,'c':30}
```

```python
# Creating Pandas series from list with default integer index

pd.Series(data=my_list)
```

```python
# Creating Pandas series from list with custom index

pd.Series(data=my_list, index=labels)
```

```python
pd.Series(my_list, labels)
```

```python
# Creating Pandas series from Array with default integer index

pd.Series(data=arr)
```

```python
# Creating Pandas series from Array with default integer index

pd.Series(arr)
```

```python
# Creating Pandas series from Array with custom integer index

pd.Series(arr, labels)
```

```python
# Creating Pandas series from Dictionary

pd.Series(dict)
```

## A pandas Series can hold a variety of object types:

```python
pd.Series(my_list, labels)
```

```python
pd.Series(labels, my_list)
```

```python
# 3. Cleaning the data set 
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#drop column 'region_code'
df.drop(columns = ['region_code'], inplace = True)
```

```python
df.head()
```

```python
# drop the 'region','covid_deaths' columns 

df.drop(columns = ['region','covid_deaths'], inplace = True)
```

```python
df.head()
```

```python
#summarize by a column
# Find the total no of deaths per country 

df.groupby('country')['total_deaths'].sum()
```

```python
df.head()
```

```python
df.groupby('country')['covid_deaths_per_100k'].sum()
```

```python
# - Find country with highest death rate

df.groupby('country')['covid_deaths_per_100k'].mean().sort_values(ascending = False).head(1)
```

```python
# - Find country with lowest death rate

df.groupby('country')['covid_deaths_per_100k'].mean().sort_values(ascending = True).head(1)
```

```python
# - Analyze the trend for a specific country over time

df_country = df[df['country'] == 'Australia']
df_country.groupby('year')['total_deaths'].sum()
```

```python
df.head()
```

```python
# 5. Visualize the result 

import matplotlib.pyplot as plt
x= df['country']
y=df['covid_deaths_per_100k']
plt.figure(figsize=(50, 15)) 
plt.xlabel('country') 
plt.ylabel('COVID Deaths')
plt.plot(x,y)
```
