```python
import numpy as np
import pandas as pd

def mpr(state):
    sheet_id = d_state_gsheet[state]
    url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv"
    df = pd.read_csv(url)
    status = 'received'
    df1 = df[df['status'] == status]

    print(f"{d_state_nm[state]} :")

    for season,season_nm in d_state_season_nm[state].items():
        print(f"{season_nm}")
        for ST,ST_nm in d_ST.items():
            if ST == 1:
                central_total = 0
                for ro in d_state_ro[state]:
                    central_total += mpr_ro(df1, season, ST, ro)
                print(f"    Central Total : {central_total}")
            elif ST == 2:
                state_total = mpr_state(df1, season, ST)
                print(f"    State Total : {state_total}")


def mpr_ro(df, season, ST, ro):
    dist_list = d_ro_distt[ro]
    count = df[(df['SEASONCODE'] == season)
                & (df['SAMPLE'] == ST)
                & (df['DISTRICTCODE'].isin(dist_list))].shape[0]

    print(f"      RO {d_ro_nm[ro]} : {count}")
    return count

def mpr_state(df, season, ST):
    count = df[(df['SEASONCODE'] == season)
                & (df['SAMPLE'] == ST)].shape[0]
    return count
```
