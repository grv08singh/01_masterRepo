```python
# finding state-wise ROs
df = (
    pd
    .read_csv(r"ALL State ML.csv", encoding='utf-8')
    .apply(lambda x : x.str.title() if x.dtype=='object' else x)
    .drop(columns = ['YEAR', 'HSEASONNAME', 'SAMPLE',
       'SAMPLENAME', 'HSAMPLENAME', 'SHORTSTATE',
       'HSTATENAME', 'HRONAME', 'SROCODE', 'SRONAME',
       'HSRONAME', 'HDISTRICTNAME', 'SELORDER',
       'EXPT', 'CROPCODE', 'CROPNAME', 'HCROPNAME'])
    .drop_duplicates()
    .astype({
        'STATE' : 'int16'
        ,'ROCODE' : 'int32'
        ,'DISTRICT' : 'int16'
    })
    .reset_index(drop=True)
    .sort_values(['STATE','ROCODE','DISTRICT'])
)
```

```python
df_state_distt_name = df.drop(columns=['SEASONCODE', 'SEASONNAME', 'STATENAME','ROCODE', 'RONAME']).drop_duplicates().sort_values(['STATE','DISTRICT']).reset_index(drop=True)
```

```python
dict_all_state_distt = {}
for state in df_state_distt_name['STATE'].unique():
    df5 = df_state_distt_name[df_state_distt_name['STATE'] == state].drop(columns=['STATE'])
    dict_distt_in_state = {}
    for distt in df5['DISTRICT']:
        dict_distt_in_state[distt] = df5[df5['DISTRICT'] == distt]['DISTRICTNAME'].values[0]
    dict_all_state_distt[state] = dict_distt_in_state
dict_all_state_distt
```
