```python
#x -> x_train,x_test
#split: train, test
#80-20,75-25,90-10,70-30#range(0,1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Standarisation

from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_scaled=sc.fit_transform(x_train)
#fit-> calculates the value to be transformed,(mean,std)
#transform-> apply the calculated value on the data

x_test_scaled=sc.transform(x_test)
```

```python
model=LogisticRegression()
```

```python
#model training

model.fit(x_train_scaled,y_train)
```

```python
#model evaluations
```

```python
#predictions
y_pred=model.predict(x_test_scaled)
```

```python
#        predicted
#           0    1
# actual 0 [TN][FP]
#        1 [FN][TP]

# TN : actual 0 and pred 0
# TP : actual 1 and pred 1
# FP : actual 0 and pred 1
# FN: actual 1 and pred 0
```

```python
0:No negative
1:Yes positive
```

```python
# accuracy : gives us overall correct prediction with respect to total predictions

# accuracy= (TP+TN)/(TN+TP+FP+FN)
(6640+342)/(6640+342+164+501)
```

```python
#precision:

#wrt 0(N):out of all -ve predicted values, how many were actually -ve
#precision:(TN/TN+FN)
precision_0=6640/(6640+501)
print(precision_0)


#wrt 1(P):out of all +ve predicted values, how many were actually +ve
#precision=(TP/TP+FP)
precision_1=342/(342+164)
print(precision_1)
```

```python
#Recall(sensitivity)

#wrt 0(N):out of all actual -ve values, how many values are correctly predicted
#recall:TN/(TN+FP)

#wrt 1(P):out of all actual +ve values, how many values are correctly predicted
#recall:TP/(TP+FN)
```

```python
# f1-score: harmonic mean of precision and recall
#  f1_score=  2/((1/Precision)+(1/Recall))
```
