Operators
1. Arithematic - +,-,*,/,%,**,//
2. Assignment op -- =,+=,-=,*=,/=,%=,**=,//=
3. Comparision op - <,>,<=,>=,!=,==
4. Logical - and, or, not

```python
a = 10
b = 5
print('Addition :',a+b)
print('Subtraction :',a-b)
print('Multiplication :',a*b)
print('Division :',a/b) #by default always you will get float result
print('Modulus :',a%b) #remainder of division
print('Exponent :',a**b) #a^b
print('Floor Division :',a//b)
```
