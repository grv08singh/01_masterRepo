```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        print(self.length * self.width)

    def perimeter(self):
        print(2 * (self.length + self.width))
```

```python
ob = Rectangle(4,3)
```

```python
ob.area()
```

```python
ob.perimeter()
```

# **Create a calculator**
* Don't create a constructor
* Four methods: ADD, SUBS, MULTI, DIV
* Use conditional statement in DIV

```python
class Calculator:
    def ADD(self,a,b):
        print(a + b)
    def SUBS(self,a,b):
        print(a - b)
    def MULTI(self,a,b):
        print(a * b)
    def DIV(self,a,b):
        if b==0:
            print('divide by zero not allowed')
        else:
            print(a / b)
```

```python
ob1 = Calculator()
ob1.ADD(210,7)
ob1.SUBS(210,7)
ob1.MULTI(210,7)
ob1.DIV(210,7)
ob1.DIV(210,0)
```

**Question 3**

Create a student class where the student will have name and marks **(a list of number(int))**. Find the average marks that calculate and **return** the average of the student

```python
class Student:
    def __init__(self, name, mark):
        self.name = name
        self.mark = mark

    def avg_marks(self):
        return sum(self.mark)/len(self.mark)
```

```python
ob2 = Student('Ajay', [98, 80, 45, 89, 65])
ob2.avg_marks()
```

**Question 4**

Create a class BankAccount where you will perform deposit, withdrawal and check_balance.

* 3 methods: Deposit, Withdrawal (It should not be greater than current balance), check_balance.
* 1 attribute: money(default = 0)

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, dep_amount):
        self.balance += dep_amount
        print('Money has been deposited.')
        print(f'your balance is {self.balance}.')
        
    def withdraw(self, w_amount):
        if w_amount > self.balance:
            print('Not enough balance.')
        else:
            self.balance -= w_amount
            print('Money has been withdrawn.')
            print(f'your balance is {self.balance}.')
            
    def check_balance(self):
        print(f'your balance is {self.balance}.')
```

```python
ob3 = BankAccount(5000)
```

```python
ob3.check_balance()
```

```python
ob3.deposit(2000)
```

```python
ob3.withdraw(3000)
```

```python
ob3.withdraw(5000)
```

```python
ob3.withdraw(4001)
```

```python
ob3.withdraw(4000)
```

# **Question 5**

Create a cla named Walmart
* attributes: Item, Price
* methods: Discount(how much), item_infos 

```python
class Walmart:
    def __init__(self, item, price):
        self.item = item
        self.price = price

    def discount(self, disc):
        self.disc = self.price - (disc * self.price / 100)

    def item_info(self):
        print(f'item: {self.item}, marked price: {self.price}, selling price: {self.disc}')
```

```python
ob4 = Walmart('Pen', 100)
```

```python
ob4.item_info()
```

```python
ob4.discount(15)
```

# **Question 6**

* Create a class NumberGuessingGame
* two private attributes: random secret_Number, attempts made
* one method: guess the secret number, if guess is > secret number then print guess lower and vice versa. if match, print secret number and the number of attempts made to reach that number.

```python
import random
class NumberGuessingGame:
    def __init__(self):
        self.__secret_Number = random.randint(1,100)
        self.__attempts = 0

    def guess(self):
        number = -99
        while number != self.__secret_Number:
            number = int(input('Enter the number: '))
            self.__attempts += 1
            if number == self.__secret_Number:
                print('correct guess!')
                print(f'the secret number is: {self.__secret_Number}, number of attempts: {self.__attempts}')
            elif number > self.__secret_Number:
                print('Incorrect number, Guess lower!')
            else:
                print('Incorrect number, Guess higher!')
```

```python
ob = NumberGuessingGame()
```

```python
ob.guess()
```
