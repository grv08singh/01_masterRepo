**Question -1**

Create the class Rectangle
* Two attribute: Length and width (Use constructor)
* Two methods: Area and Perimeter of rectangle

```python
class Rectangle:
  def __init__(self,l,w):
    self.length = l
    self.width = w

  def area(self):
    print(f'The area of Rectangle is {self.length * self.width}')

  def perimeter(self):
    print(f'The perimeter of Rectangle is {2 * (self.length + self.width)}')
```

```python
ob1 = Rectangle(27,16)
ob1.area()
ob1.perimeter()
```

**Question -2**

Create the class Calculator
* **Do not create contructor**
* Four methods: ADD,SUBS,MULTI,DIV
* Use conditional statement in DIV method

```python
class Calculator:
  def add(self,a,b):
    print(a+b) # self.a +self.b -- If we create attribute
  def subs(self,a,b):
    print(a-b)
  def multi(self,a,b):
    print(a*b)
  def div(self,a,b):
    if b!=0:
      print(a/b)
    else:
      print('You cannot divide a number with 0')
```

```python
ob = Calculator()
ob.add(10,20)
ob.subs(10,20)
ob.multi(5,2)
ob.div(10,5)
ob.div(10,0)
```

Create a student class where the student will have name and marks of 5 subjects(a list of number(int)). Find the average marks that calculate and **return** the average of the student.

**hint**
* Use len() and sum() function

```python
class Student:
  def __init__(self,name,grade):
    self.name = name
    self.grade = grade

  def averageGrade(self):
    return f'Name: {self.name}, average grade: {sum(self.grade)/len(self.grade)}'
```

**Question - 4**

Create a BankAccount where you will perfrom Deposit, Withdraw and check balance.

* 3 methods: Deposit, withdraw(It should not be greater than current balance), Check balance
* 1 Attrubute: Money(default = 0)

```python
class BankAccount:
  def __init__(self,money = 0):
    self.money = money

  def deposit(self,amt):
    self.money += amt
    return f'Updated Balance: {self.money}'

  def withdraw(self,amt):
    if amt<=self.money:
      self.money -=amt
      print(f'Updated Balance: {self.money}')
    else:
      print('Insufficient balance')

  def checkBalance(self):
    print(f'Current balance: {self.money}')
```

```python
acc1 = BankAccount(1000)
acc1.checkBalance()
print(acc1.deposit(2000))
acc1.withdraw(5000)
print(acc1.deposit(3000))
acc1.withdraw(6000)
acc1.checkBalance()
```

**Question -5**

Create a class named Walmart:
* Two attributes: Item, Price(use constructor)
* Two methods: Discount(howMuch) and item_info

```python
class Walmart:
  def __init__(self,item,price):
    self.item = item
    self.price = price

  def discount(self,dis):
   self.disc =  self.price - (self.price*(dis/100))

  def item_info(self):
    print(f'MRP of {self.item} is {self.price}. After Discount price is {self.disc}')
```

```python
ob = Walmart('Chair',800)
ob.discount(10)
ob.item_info()
```

**Question -6**
* class NumberGuessingGame
* Constructor = default value---> Secret number, attempts (these attributes should be hidden)
* Create one method - guess(number)
    * If number < secret_number --- Guess higher-- Attempt +=1
* Create object through: while loop

```python
import random
class NumberGuessingGame:
  def __init__(self):
    self.__secretnum = random.randint(1,100)
    self.__attempts = 0

  def guess(self,number):
    self.__attempts +=1
    if number < self.__secretnum:
      return 'Incorrect Guess -- Go Higher'
    elif number > self.__secretnum:
      return 'Incorrect Guess -- Go Lower'
    else:
      return f'Congo..you won in {self.__attempts} attempts'
```

```python
ob = NumberGuessingGame()
ob.guess()
while True: #------ Break statement
  x = int(input('Guess the number between 1-100: '))
  answer = ob.guess(x)
  print(answer)
  if 'Congo' in answer:
    break
```
