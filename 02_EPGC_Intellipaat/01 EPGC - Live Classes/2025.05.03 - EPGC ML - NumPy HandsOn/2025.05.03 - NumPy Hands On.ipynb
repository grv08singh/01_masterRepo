# enumerate

```python
fruits = ['apple', 'banana', 'orange']
for i,v in enumerate(fruits):
    print(i, v)
```

```python
fruits = ['apple', 'banana', 'orange']
for i,v in enumerate(fruits, start=1):
    print(i, v)
```

```python
scores = np.array([65, 70, 75, 80, 85, 90, 95])
# normalize the score to a 0 - 1 scale using formula 
# normalize = (x-min) / (max- min)
```

```python
score1 = (scores - min(scores)) / (max(scores) - min(scores))
print(score1)
```

```python
score2 = (scores - scores.min()) / (scores.max() - scores.min())
print(score2)
```

```python
person_a = np.array([1200, 1100, 1300, 1250, 1400, 1350])
person_b = np.array([1000, 1150, 1200, 1100, 1250, 1300])
# find out in which month person a spent more than person b

for i in range(len(person_a)):
    if person_a[i]>person_b[i]:
        print(f'month {i+1}')
```

```python
result = person_a > person_b
months = np.where(result)[0]+1
print(months)
```
