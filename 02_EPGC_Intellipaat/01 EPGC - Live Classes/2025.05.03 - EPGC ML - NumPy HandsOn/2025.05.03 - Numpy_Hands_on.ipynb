```python
import numpy as np

# Zero-dimensional array
zero_dim_array = np.array(42)
print("Zero-dimensional array shape:", zero_dim_array.shape)
print("Zero-dimensional array dimensions:", zero_dim_array.ndim)

# One-dimensional array
one_dim_array = np.array([1, 2, 3, 4, 5])
print("One-dimensional array shape:", one_dim_array.shape)
print("One-dimensional array dimensions:", one_dim_array.ndim)

# Two-dimensional array
two_dim_array = np.array([[1, 2, 3], [4, 5, 6]])
print("Two-dimensional array shape:", two_dim_array.shape)
print("Two-dimensional array dimensions:", two_dim_array.ndim)

# Three-dimensional array
three_dim_array = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("Three-dimensional array shape:", three_dim_array.shape)
print("Three-dimensional array dimensions:", three_dim_array.ndim)

# Four-dimensional array
four_dim_array = np.array([[[[1], [2]], [[3], [4]]], [[[5], [6]], [[7], [8]]]])
print("Four-dimensional array shape:", four_dim_array.shape)
print("Four-dimensional array dimensions:", four_dim_array)
```

```python
sales = np.array([250,  270 , 260 , 5000, 280, 290 , 275])


# detect the sales greater than 1000 as outlier and replace them with the mean of rest


non_outlier =  sales[sales <=1000]
mean_value = np.mean(non_outlier)
print(mean_value)

sales[sales> 1000] = mean_value
print(sales)
```

```python
scores = np.array([65, 70, 75, 80, 85, 90, 95])

# normalize the score to a 0 - 1 scale using formula
# normalize = (x-min) / (max- min)

# 5 mins

min = scores.min()
max = scores.max()
normalize = (scores - min) / (max - min)
normalize
```

```python
scores = np.array([65, 70, 75, 80, 85, 90, 95])
# normalize the score to a 0 - 1 scale using formula
# normalize = (x-min) / (max- min)
normalize = []
for i in range(len(scores)):
    normalize.append((scores[i] - np.min(scores)) / (np.max(scores) - np.min(scores)))
print(normalize)
```

```python
scores_min = np.min(scores, axis=0)
scores_max = np.max(scores, axis=0)
normalized_data = (scores - scores_min) / (scores_max - scores_min)
print("Normalized data:\n", normalized_data)
```

```python
hours = np.array([40, 38, 42, 36, 45])
# find out the average working hours and also list those employees who worked more
# than average hours

avg_hours = np.mean(hours)
above_avg = hours> avg_hours
print("average hours", avg_hours)
print("employee with above average hours", np.where(above_avg)[0]+1)
```
