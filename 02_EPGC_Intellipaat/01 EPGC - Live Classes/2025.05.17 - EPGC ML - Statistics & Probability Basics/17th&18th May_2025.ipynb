```python
d.append(300)
```

```python
plt.boxplot(d)
```

```python
q1 = np.percentile(d,25)
q3 = np.percentile(d,75)
```

```python
print(q1,q3)
```

```python
iqr = q3-q1 #inter quartile range
print(iqr)
```

```python
ll = q1 - 1.5*iqr
ul = 1.5*iqr + q3
print(ll,ul)
```

```python
d = [x for x in d if ll<= x <= ul]
```

```python
d
```

```python
d1 = []
for x in d:
  if ll<= x <= ul:
    d1.append(x)
```
