Q4. Create a function to check if a number is prime or not?

```python
def prime_checker(num):
  if num<=1 :
    return 'Not Prime'
  elif num == 2:
    return 'Prime'
  else:
    for i in range(2,num):
      if num%i==0:
        return 'Not Prime'
      else:
        return 'Prime'
```

```python
prime_checker(7)
```

```python
prime_checker(8)
```

Q5. Create a function to check if a string is palindrome or not

```python
def Palindrome(st):
  st = st.lower()
  if st == st[::-1]: #start:stop:step
    return 'Palindrome'
  else:
    return 'Not Palindrome'
```

```python
Palindrome('eye')
```

```python
Palindrome('Eye')
```

Q6. Create a function to calculate the factorial of a number

```python
def Factorial(num):
  fact = 1
  for i in range(1,num+1):
    fact*=i
  return fact
```

Q9. Create a function to check if 2 strings are anagram or not?

```python
def Anagram(st1,st2):
  st1 = st1.lower()
  st2 = st2.lower()
  if sorted(st1)==sorted(st2):
    return 'Anagram'
  else:
    return 'Not Anagram'
```

```python
Anagram('poonam','anoopm')
```
