```python
#Q_04) Create a function to check whether a number is prime or not.
import math
def prime_checker(a):
    flag = 1
    for i in range(2, 2+int(math.sqrt(a))):
        if a%i == 0:
            flag = 0
            break
    if a == 1:
        print('Neither Prime nor Composite.')
    elif a == 2 or flag == 1:
        print('Prime')
    else:
        print('Not Prime')
        
```

```python
prime_checker(2)
prime_checker(3)
prime_checker(4)
prime_checker(5)
prime_checker(6)
prime_checker(7)
prime_checker(8)
prime_checker(9)
prime_checker(10)
```

```python
#Q_05) Create a function to check if a string is pallindrome or not.
def pallindrome(s):
    if s.lower() == s[::-1].lower():
        return 'Pallindrome'
    else:
        return 'Not a Pallindrome'
```

```python
pallindrome('Madam')
```

```python
pallindrome('car')
```

```python
pallindrome('rickshaw')
```

```python
#Q_06) Create a function to create the factorial of a number.
def factorial(n):
    fact = 1
    for i in range(n,1,-1):
        fact *= i
    return fact
```

```python
factorial(3)
```

```python
factorial(4)
```

```python
factorial(5)
```

```python
factorial(6)
```

```python
#Q_09) Create a function to check if two strings are anagram or not?
def anagram(s1, s2):
    if len(s1) == len(s2):
        s1 = s1.lower()
        s2 = s2.lower()
        if sorted(s1) == sorted(s2):
            return True
        else:
            return False
    else:
        return False
```

```python
anagram('truue','rutee')
```

```python
#Q_10) Calculate the sum of digits of a number.
a = 123456789
if len(str(a)) > 1:
    digit_sum = a
    while len(str(digit_sum)) > 1:
        temp = 0
        for i in str(digit_sum):
            temp += int(i)
        digit_sum = temp
    print(digit_sum)
else:
    print(a)
```
