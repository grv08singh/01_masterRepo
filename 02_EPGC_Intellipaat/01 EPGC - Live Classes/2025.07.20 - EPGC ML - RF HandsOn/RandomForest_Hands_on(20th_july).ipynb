```python
# use different values of n _estimator to find which optimal no of dt if going to give us highest accuracy

n = [50,100,200,300,400,500,600,700,800,900,1000]

accuarcy= []

for i in n:
  model = RandomForestClassifier(n_estimators= i)
  model.fit(X_train, y_train)
  ans = model.predict(X_test)
  a = accuracy_score(y_test ,ans)*100
  accuarcy.append(a)
```

```python
accuarcy

d = {'number of tree': n, "accuracy of the in RF": accuarcy}
df = pd.DataFrame(d)
print(df)
```

```python
y = churn_data['Churn']
X1 = churn_data.iloc[:, :-1]
```

```python
oobclassifier =RandomForestClassifier(n_estimators=800, oob_score=True, random_state=0)
```

```python
oobclassifier.fit(X1, y)
```

```python
oob_data_indices = np.where(oobclassifier.oob_decision_function_[:,1]>= 0.5)

 # give the probabiliyt of churn == yes for each sample
 # we are selecting the indices of thoe data points where the model predict churn probability >  = 50 %
```

```python
oob_dataset = X1.iloc[oob_data_indices]

oob_dataset
```

```python
print(oobclassifier.oob_score_)
```
