### Convert the `raw digital image` to `Std. RGB` image

```python
img = img.convert('RGB')
```

```python
img
```

### Convert the above image to `GrayScale`

```python
img_gray = img.convert('L') # L stands for Luminance
```

```python
img_gray
```

```python
### Get the size of the image
width, height = img_gray.size #it returns a tuple (width, height) of the image in pixels
```

```python
print("Width of the number 7 is:", width, "px")
print("Height of the number 7 is:", height, "px")
```

```python
### Convert this image from 144 X 155 to 28 X 28
img_gray_resized = img_gray.resize((28,28))
```

```python
### Get the size of the image
width1, height1 = img_gray_resized.size #it returns a tuple (width, height) of the image in pixels
```

```python
print("Width of the resized number 7 is:", width1, "px")
print("Height of the resized number 7 is:", height1, "px")
```

```python
img_gray_resized
```

```python
### Convert the resized grayscale image into a pixelated array
img_gray_resized_array = np.array(img_gray_resized)
```

```python
img_gray_resized_array
```

```python
### Plotting the `pixelated image` as a 28 by 28 grid
plt.figure(figsize = (4,4))
plt.imshow(img_gray_resized_array, cmap='gray')
plt.colorbar()
plt.show()
```
