```python
for i in df.columns:
  if df[i].dtype in ['float64']:
    plt.boxplot(df[i])
    plt.xlabel(i)
    plt.show()
```

```python
for i in df.columns:
  if df[i].dtype in ['float64']:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3-Q1
    LB = Q1-1.5*IQR
    UB = Q3+1.5*IQR
    df = df[(df[i] >= LB)& (df[i] <= UB) ]
```

```python
for i in df.columns:
  if df[i].dtype in ['float64']:
    plt.boxplot(df[i])
    plt.xlabel(i)
    plt.show()
```
