```python
import matplotlib.pyplot as plt 
import numpy as np 

# data 
a = np.linspace(0,10,20)
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
marks = [85, 92, 78, 88, 95]
fruits = ['Mango', 'Apple', 'Banana', 'Grapes', 'Orange']
prices = [10, 20, 30, 40, 50]
votes = [-100, 2, 4, 6, 8, 99, 200, 56, 45, 22, 45]
x = np.random.randint(0, 100, 20)
y = np.random.randint(0, 100, 20)
c_name = ["TCS", 'TATA','BPCL', 'HCL', 'ABC Inc']
profit = [10, 20,30, 40, 50]

# figure size 
plt.figure(figsize = (20,10))

# axes
ax1 = plt.subplot(231)
plt.xlabel('A')
plt.ylabel('A')
plt.title('A')

ax2 = plt.subplot(232)
plt.xlabel('A')
plt.ylabel('A')
plt.title('A')
ax3 = plt.subplot(233)
plt.xlabel('A')
plt.ylabel('A')
plt.title('A')
ax4 = plt.subplot(234)
plt.xlabel('A')
plt.ylabel('A')
plt.title('A')
ax5 = plt.subplot(235)
plt.xlabel('A')
plt.ylabel('A')
plt.title('A')
ax6 = plt.subplot(236)
plt.xlabel('A')
plt.ylabel('A')
plt.title('A')

ax1.plot(a)
ax2.bar(students, marks, color='skyblue')
ax3.barh(fruits, prices)
ax4.boxplot(votes)
ax5.scatter(x, y)
explode = (0, 0, 0, 0,0)
ax6.pie(profit, labels=c_name, explode=explode, autopct = '%1.2f%%')

plt.show()


```
