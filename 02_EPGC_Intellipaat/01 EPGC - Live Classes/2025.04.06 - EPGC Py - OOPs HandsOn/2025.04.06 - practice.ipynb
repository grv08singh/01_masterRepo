```python
def armstrong(n):
    sum = 0
    power = len(str(n))
    for c in str(n):
        sum += int(c)**power
    if n == sum:
        return 'Yes, Armstrong Number.'
    else:
        return 'No, Not an Armstrong Number'
```

```python
armstrong(10)
```

```python
num = int(input('Enter the number: '))
count = len(str(num))
temp = num
sum = 0
while temp > 0:
  uniDigit = temp % 10
  sum = sum + uniDigit**count
  temp = temp//10
if num == sum:
  print('Armstrong')
else:
  print('Not Armstrong')
```
