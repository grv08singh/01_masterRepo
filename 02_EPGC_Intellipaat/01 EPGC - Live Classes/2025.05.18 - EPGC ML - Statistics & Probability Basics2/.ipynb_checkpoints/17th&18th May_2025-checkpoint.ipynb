```python
d = [1,2,3,4,5,6,7,8,10]
```

```python
np.percentile(d,100)
```

```python
np.percentile(d,50)
```

```python
np.percentile(d,0)
```

```python
#Percentile means the value X below with that much x% of data falls
#depends upon the no of participants
```

```python
np.percentile(d,25)
```

```python
np.percentile(d,75)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(d)
```

```python
d.append(300)
```

```python
plt.boxplot(d)
```

```python
q1 = np.percentile(d,25)
q3 = np.percentile(d,75)
```

```python
print(q1,q3)
```

```python
iqr = q3-q1 #inter quartile range
print(iqr)
```

```python
ll = q1 - 1.5*iqr
ul = 1.5*iqr + q3
print(ll,ul)
```

```python
d = [x for x in d if ll<= x <= ul]
```

```python
d
```

```python
d1 = []
for x in d:
  if ll<= x <= ul:
    d1.append(x)
```

```python
d1
```
