```python
#Create one employee management system. Create a class Employee with attributes name,empid,year_of_joining. You have to create 2 methods 
#1. promotion --> take th input of age and check if the employee is moree than 30 years and joined before 2022, give promotion and show his details else no promotion
#2. Salary --> Take the input of his allowances HA,TA,MA and Basic pay and calculate and display his salary
```

```python
class employee:
    def __init__(self, name, empid, yoj):
        self.name = name #class variable
        self.empid = empid #class variable
        self.yoj = yoj #class variable

    def promotion(self, age):
        if age > 30 and self.yoj < 2022: #age: method variable (not class variable)
            print('Promoted')
        else:
            print('Not Promoted')
        print(f'name = {self.name}, empid = {self.empid}, year of joining = {self.yoj}')

    def salary(self, ha, ta, ma, bp): #all method variables
        print(f'Total Salary = {ha+ta+ma+bp}')
```

```python
obj = employee('Ashu', 112, 2021)
```

```python
obj.promotion(30)
```

```python
obj.salary(100,200,300,500)
```

```python
#Create a Interest calculator class with 3 attributes principal,rate_of_int,year, 
#create 2 methods simple_interest() and compound_interest() and 
#display the total interest on a pricipal comout
```

```python
class int_calc:
    def __init__(self, p, r, t):
        self.p = p
        self.r = r
        self.t = t
    def simp_int(self):
        si = self.p*self.r*self.t/100
        return int(si)
    def comp_int(self):
        ci = self.p*((1 + self.r/100)**self.t - 1)
        return int(ci)
```

```python
ic = int_calc(1000,10,3)
```

```python
ic.simp_int()
```

```python
ic.comp_int()
```
