Attributes - variables which are accesssible inside the entire class in any method

Self parameter - mandatory parameter in all methods inside a class which gives access to the variables to be used in a method

Constructor - it is a special method used inside a class to initialize attributes in the class

```python
class Calculator:
  def __init__(self,a,b):
    print(a,b)
    self.a = a #attributes
    self.b = b #global to the class
  def add(self,name):
    print(name)
    return self.a + self.b
  def subtract(self):
    return self.a-self.b
  def multiply(self):
    return self.a*self.b
  def divide(self):
    return self.a/self.b
  def modulus(self):
    return self.a%self.b
```

```python
o = Calculator(10,4)
```

```python
o.add('Poonam')
```

```python
o.subtract()
```

```python
o.multiply()
```

```python
o.divide()
```

```python
o.modulus()
```

Create one employee management system. Create a class Employee with attributes name,empid,year_of_joining. You have to create 2 methods

1. promotion --> take th input of age and check if the employee is moree than 30 years and joined before 2022, give promotion and show his details else no promotion

2. Salary --> Take the input of his allowances HA,TA,MA and Basic pay and calculate and display his salary

```python
class Employee:
  def __init__(self,name,empid,yoj):
    self.name = name
    self.empid = empid
    self.yoj = yoj
  def promotion(self):
    age = int(input('Enter your age :'))
    if age > 30 and self.yoj <2022:
      return f"Eligible for Promotion for {self.name},{self.empid}"
    else :
      return 'No promotion'
  def salary(self):
    HA = int(input('Enter HA :'))
    TA = int(input('Enter TA :'))
    MA = int(input('Enter MA :'))
    BP = int(input('Enter BP :'))
    sal = HA+TA+MA+BP
    return sal
```

```python
o = Employee('poonam',123,2021)
```

```python
o.promotion()
```

#Pillars of OOPs
1. Inheritance
2. Polymorphism
3. Encapsulation
4. Abstraction

```python
#Inheritance
# 1. Single - 1 parent, 1 Child
# 2. Multiple - many parents , 1 Child
# 3. Multi-level - level wise inheritance GF --> F --> C
# 4. Hierarchical - 1 parent, many children
# 5. Hybrid - combination of all
```

```python
#single -
class Parent :
  def display_name(self):
    return 'Poonam'


class Child(Parent):
  def display_company(self):
    return 'Intellipaat'
```

```python
o = Child()
o.display_name() #parent class method called using child obj
```

```python
o.display_company()
```

```python
#with const
class ParentNew :
  def __init__(self,name,city):
    self.name = name
    self.city = city
  def show(self):
    print(self.name,self.city)


class ChildNew(ParentNew):
  def __init__(self,name,city,state):
    super().__init__(name,city) #calls constructor method from parent class
    self.state = state
  def showall(self):
    print(self.state)
    super().show() #call any parent class method and use it in child class
```

```python
o = ChildNew('Poonam','Bengaluru','Karnataka')
```

```python
o.show()
```

```python
o.showall()
```
