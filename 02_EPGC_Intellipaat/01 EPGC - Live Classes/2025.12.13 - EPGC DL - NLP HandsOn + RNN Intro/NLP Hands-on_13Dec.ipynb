### 3 & 4. `Stemming` & `Lemmatization`

- Stemming using porter

```python
from nltk.stem import PorterStemmer
import time
```

```python
words = ["running", "flies", "easily", "fairly", "happiness", "arguing", "automation", "studies", "organization", "studying", "everyone", "transforming",
         "stunning", "better", "antagonizing", "good", "goods", "goodies", "liking", "likes",  "hurry","hurried","begin", "beginning", "begins", "unwell", "unhappy", "best"]
```

```python
stemmer = PorterStemmer()
```

```python
### Start Timing
start_time = time.time()


### Run stemming
stemmed = [stemmer.stem(word) for word in words]

end_time = time.time()
time_elapsed = (end_time - start_time)
```

```python
print("Original Words:", words,"\n")
print("Stemmed Words:", stemmed, "\n")
print(f"Time taken for stemming: {time_elapsed : .3f} sec") 
```

```python
plt.figure(figsize=(12, 6))

# Define custom colors for each sentiment
sentiment_colors = ['red', 'orange', 'green']  # negative, neutral, positive

# Create a crosstab of airline and sentiment
ct = pd.crosstab(data['airline'], data['airline_sentiment'])

# Plot the stacked bar chart with custom colors
ax = ct.plot(kind='bar', stacked=True, ax=plt.gca(), color=sentiment_colors)

plt.title('Percentage of Tweets by Airline and Sentiment')
plt.xlabel('Airline')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45)
plt.legend(title='Sentiment', labels=ct.columns)

# Annotate each segment with count and percentage
for i, airline in enumerate(ct.index):
    total = ct.loc[airline].sum()
    bottom = 0
    for j, sentiment in enumerate(ct.columns):
        count = ct.loc[airline, sentiment]
        percent = 100 * count / total if total > 0 else 0
        if count > 0:
            ax.text(
                i, 
                bottom + count / 2, 
                f'{count}\n({percent:.1f}%)', 
                ha='center', va='center', fontsize=10, color='black'
            )
        bottom += count

plt.tight_layout()
plt.show()
```

### Replace Contractions in string (e.g., replace `I'm --> I am`, `We'll --> We will`) and so on

- Contractions are shortened versions of words or phrases, common in both spoken and written language. They are formed by combining two words and replacing some of the letters with an apostrophe. Their primary purpose is to make language more conversational and efficient.


```python
def replace_contractions(text):
    '''
    Replace contractions in strings of text
    didn't --> did not
    '''
    return contractions.fix(text)
```

```python
lemmatizer = WordNetLemmatizer()

def remove_non_ascii(words):
    """Remove non-ASCII characters from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = unicodedata.normalize('NFKD', word).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        new_words.append(new_word)
    return new_words

def to_lowercase(words):
    """Convert all characters to lowercase from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = word.lower()
        new_words.append(new_word)
    return new_words

def remove_punctuation(words):
    """Remove punctuation from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = re.sub(r'[^\w\s]', '', word)
        if new_word != '':
            new_words.append(new_word)
    return new_words

def remove_stopwords(words):
    """Remove stop words from list of tokenized words"""
    new_words = []
    for word in words:
        if word not in stopwords_set:
            new_words.append(word)
    return new_words

def lemmatize_list(words):
    new_words = []
    for word in words:
      new_words.append(lemmatizer.lemmatize(word, pos='v'))
    return new_words
```

```python
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
```

```python
def normalize(words):
    words = remove_non_ascii(words)
    words = to_lowercase(words)
    words = remove_punctuation(words)
    words = remove_stopwords(words)
    words = lemmatize_list(words)
    return ''.join(words)

data['text'] = data.apply(lambda row: normalize(row['text']), axis=1)
    
```

### Build the model based on `CountVectorizer` and `RandomForest`

```python
from sklearn.feature_extraction.text import CountVectorizer #to convert a collection of text docs into matrix of token counts
```

```python
Count_vec = CountVectorizer(max_features=5000) #keeping only 5000 words or dictionary words hence only 5000 features
data_features = Count_vec.fit_transform(data['text'])
data_features_array = data_features.toarray()
```

```python
X = data_features_array #training features (input)
y = data.airline_sentiment #training labels (target/output)
```

```python
y
```

```python
### Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Using `random forest` to build the model for classification of tweets' sentiment

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

nlp_rf = RandomForestClassifier(n_estimators=50, n_jobs=6)
'''
n_estimators => specifies the number of decision trees in the forest
n_jobs => means that how many CPU cores will be used to parallize the computation --> which speeds up the training process
'''
### Training the model
nlp_rf = nlp_rf.fit(X_train, y_train)
```

```python
print(np.mean(cross_val_score(nlp_rf, X, y, cv=5))) #training accuracy
```

```python
### Generate accuracy on the test set
print(nlp_rf.score(X_test, y_test)) #testing accuracy
```
