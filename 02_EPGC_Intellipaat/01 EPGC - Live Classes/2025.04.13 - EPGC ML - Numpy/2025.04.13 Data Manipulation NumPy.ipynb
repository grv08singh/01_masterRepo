# ** numpy - Data Manipulation - HandsOn**

```python
import numpy as np 
```

```python
# create an array with the help of list having values as float
```

```python
a1 = np.array([1,2,3,4,5,6], float)
```

```python
print(a1)
```

```python
# create an array with the help of zeros with 5 rows and 2 columns
```

```python
a2 = np.zeros((5,2))
```

```python
print(a2)
```

```python
# create an array with the help of full with values as 33 
```

```python
a3 = np.full((5,7), 33)
```

```python
print(a3)
```

```python
# create an array with the help of ones with 9 rows and 3 columns
```

```python
a4 = np.ones((9,3))
```

```python
print(a4)
```

```python
# create an array with the help of linspace with equal spacing of 2 
```

```python
a5 = np.linspace(10,30,11)
```

```python
print(a5)
```

```python
# create an array with the help of arange with step size as 3
```

```python
a6 = np.arange(0,30,3)
```

```python
print(a6)
```

```python
# create an array with the help of random in which values lie between 10 and 11
```

```python
a7 = np.random.random(5)+10
```

```python
print(a7)
```

```python
# create an array with the help of randomint between 10 and 100
```

```python
b = []
for i in range(5):
    b.append(np.random.randint(10,100))
a8 = np.array(b)
```

```python
print(a8)
```

```python
# create a 3d array and convert it into 2d and than into 1 d
```

```python
a9 = np.array([[[4,6,3,2],[1,4,2,5],[1,9,0,7]],[[3,5,6,4],[0,0,3,4],[2,7,7,5]]])
print(a9)
a10 = a9.reshape(4,6)
print(a10)
a11 = a9.reshape(24)
print(a11)
```
