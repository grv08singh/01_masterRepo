```python
# Create a fucntion which will take df and list of columns containing ouliers and the function returns the df with outliers elimated
def outlier_remover(df,out_col):
  for col in out_col:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)

    IQR= Q3-Q1

    LB = Q1-1.5*(IQR)
    UB= Q3+ 1.5*(IQR)

    df=df[(df[col]>=LB) & (df[col]<=UB)]

  return df
```

```python
# 4. Outliers
def outlier_remover(df):
    for col in df.columns:
        if(df[col].dtype != 'object'):
            Q1=df[col].quantile(0.25)
            Q3=df[col].quantile(0.75)
            IQR = Q3-Q1
            LB=Q1-1.5*(IQR)
            UB=Q3+1.5*(IQR)
            df=df[(df[col]>=LB) & (df[col]<=UB)]
    return df
```

```python
# 5. Label Encoding
def label(df):
  from sklearn.preprocessing import LabelEncoder
  LE=LabelEncoder()
  for col in df.columns:
    if(df[col].dtype=='object'):
      df[col]=LE.fit_transform(df[col])
  return df
```

```python
df = label(df)
```

```python
df = outlier_remover(df)
```
