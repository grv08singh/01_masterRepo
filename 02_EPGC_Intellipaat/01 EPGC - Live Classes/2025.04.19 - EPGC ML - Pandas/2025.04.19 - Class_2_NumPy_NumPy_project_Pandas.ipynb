```python
# reshape

a = np.arange(24)
print(a)
print(np.ndim(a))
print()

b = a.reshape(6,4)
print(b)
print(np.ndim(b))
print()

c = a.reshape(2,3,4)
print(c)
print(np.ndim(c))
```

# Indexing and slicing

```python
# indexing
a = np.array([10,20,30,40,50])
# 10 
a[0]
```

```python
# 40 
a[3]
```

```python
# slicing
a = np.array([10,20,30,40,50])
# [10,20]

# a[start:end+1:step]
a[0:2]
```

```python
a = np.array([10,20,30,40,50])
# [30,40,50]
a[2:5]
```

```python
a = np.array([10,20,30,40,50])
#[10,30,50]
a[0:5:2]
```

```python
# 2d 

a = np.array([[10,20,30],
              [40,50,60]])
# a[rows,columns]
# 10 
a[0,0]
```

```python
a = np.array([[10,20,30],
              [40,50,60]])
# 50 
a[1,1]
```

```python
a = np.array([[10,20,30],
              [40,50,60]])
# 30
a[0,2]
```

```python
# slicing 
a = np.array([[10,20,30],
              [40,50,60]])
# a[rstart:rend+1:rstep, cstart:cend+1:cstep]
# [10,20]
a[0:1:1,0:2:1]
```

```python
a = np.array([[10,20,30],
              [40,50,60]])

# [[10],
#  [40]]

a[0:2:1, 0:1:1]
```

```python
a = np.array([[10,20,30],
              [40,50,60]])

# [[50,60]]
# a[rstart:rend+1:rstep, cstart:cend+1:cstep]
a[1:2:1,1:3:1]
```

```python
a = np.array([[10,20,30],
              [40,50,60]])

# [[20,30]]
a[0:1:1,1:3:1]
```

```python
a = np.array([[10,20,30],
              [40,50,60]])

# [[20],
#  [50]]

a[0:2:1,1:2:1]
```

```python
a = np.array([[10,20,30],
              [40,50,60]])

# [[10,30],
# [40,60]]

a[0:2:1,0:3:2]
```

# Series

```python
- 1d dimensional
```

```python
import pandas as pd
a = pd.Series([10,20,30,40,50])
print(a)
print(type(a))
```

```python
a = pd.Series([10,20,30,40,50], index = ['a','b','c','d','e'])
print(a)
```

```python
a = pd.Series([10,20,30,40,50], index = ['a','b','c','d','e'], dtype = float)
print(a)
```

```python
a = pd.Series([[101,'Hemant', 'Google', 'DS', 500000],
              [102,'Hema', 'HCL', 'DA', 400000],
              [103,'Ant', 'Wipro', 'AwS', 300000],
              [104,'Ema', 'TCS', 'WD', 200000],
              [105,'Mant', 'INFY', 'CS', 100000]])
a
```
