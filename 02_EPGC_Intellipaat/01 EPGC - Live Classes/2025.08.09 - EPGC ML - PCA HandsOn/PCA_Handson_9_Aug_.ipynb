```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
df_scaled=sc.fit_transform(df)
df_scaled
```

```python
df_scaled.shape
```

```python
# co-variance matrix
# eigigen value and vector
# select optimal number of components

# we will try to go with diffrent values for components and also we'll try to go with a diffrent method
 #where we try to pass the percentage of information that has to be retrived while reducing the dimension
```

```python
from sklearn.decomposition import PCA
```

```python
model=PCA(n_components=35)  #either I can pass random number columns i want to reduse to OR i can pass the information i want to retrive
```

```python
# I Want to have 95% of Information still present in my transformed data after I reduce the dimension from 64 columns to X columns
```

```python
x_PCA=model.fit_transform(df_scaled)
x_PCA
```

```python
x_PCA.shape
```

```python
x_PCA=pd.DataFrame(x_PCA)
x_PCA
```

```python
y=dataset.target
```

```python
y
```

```python
# Build a ML Model and check if with these X columns I have redused to from 64 will still be able to predict the target values

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_PCA,y,test_size=0.2,random_state=234)
```

```python
from sklearn.metrics import *
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier()
```

```python
rf_model.fit(x_train,y_train)
```

```python
pred=rf_model.predict(x_test)
pred
```

```python
accuracy_score(y_test,pred)*100
```
