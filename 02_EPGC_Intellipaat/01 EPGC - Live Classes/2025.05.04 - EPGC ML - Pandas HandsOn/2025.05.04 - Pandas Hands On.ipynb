```python
#3) 3. Create a pandas series, with the elements as numbers from 1-10. 
# Use the index values starting from 0-10. For the above series:
# a. Convert it to a list.
# b. Convert it to a dictionary.
# c. Convert it to a tuple.
# d. Convert it to a set


s3 = pd.Series(range(1,11))
#a.
l = list(s3)
print(l)
#b.
d = s3.to_dict()
print(d)
#c.
t = tuple(s3)
print(t)
#d.
s = set(s3)
print(s)
```
