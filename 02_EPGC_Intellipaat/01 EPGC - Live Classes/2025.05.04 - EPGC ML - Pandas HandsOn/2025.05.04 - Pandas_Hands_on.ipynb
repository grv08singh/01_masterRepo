```python
A = pd.DataFrame({
    'Name' : ['Alpha', 'Beta', 'Gamma', 'Delta', 'Epsilon'],
    'Age' : [25, 30, 10, 40,55]
})
B = pd.DataFrame({
    'Name' : ['Zeta', 'Eta', 'Theta', 'Iota', 'Kappa'],
    'Age' : [50, 10, 60, 65, 70]
})
C = pd.concat([A, B]).reset_index(drop=True)
print(C)
print('Spliting dataframe row wise:')
C1 = C.iloc[:5, :]
C2 = C.iloc[5:, :]
print(C1)
print(C2)
print('Spliting dataframe column wise:')
C3 = C[C['Age'] > 10]
C4 = C[C['Age'] <= 10]
print('Spliting dataframe column wise: more than 10')
print(C3)
print('Spliting dataframe column wise: more less/equal 10')
print(C4)
```

Create a Data Frame with columns Name and Age. Perform the following operations.
 a. Write a sample code to search for the Name/Names where Age is 20.

 b. Sort the Data Frame in ascending order of ages.

 c. Using Python loops, print the ages from the data frame and store it in a set

```python
# 5mins
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 20, 20, 30, 22]
}
df = pd.DataFrame(data)
print(df)

names_age_20 = df[df['Age'] == 20]['Name']
print(names_age_20)

df_sorted = df.sort_values(by='Age')
print(df_sorted)

ages_set = set()
for age in df['Age']:
    print(age)
    ages_set.add(age)

print("\nSet of ages:", ages_set)
```

```python
person = pd.DataFrame({
    'Name' : ['Alpha', 'Beta', 'Gamma', 'Delta', 'Epsilon'],
    'Age' : [25, 30, 20, 40, 20]
})
print(person)
print('Search for the Name/Names where Age is 20')
print(person[person['Age'] == 20])
print('Sort the Data Frame in ascending order of ages')
print(person.sort_values(by='Age', ascending=True))
print('Using Python loops, print the ages from the data frame and store it in a set')
set1 = set()
for age in person['Age']:
  set1.add(age)
print(set1)
```

```python
import pandas as pd

# Create a Data Frame
data = {
    'Name': ['Aarav', 'Vivaan', 'Aditya', 'Vihaan', 'Arjun', 'Sai'],
    'Age': [20, 22, 20, 21, 23, 20]
}
df = pd.DataFrame(data)

# a. Search for the Name/Names where Age is 20
names_age_20 = df[df['Age'] == 20]['Name'].tolist()
print(names_age_20)

# b. Sort the Data Frame in ascending order of ages
sorted_df = df.sort_values(by='Age')
print(sorted_df)

# c. Using Python loops, print the ages from the data frame and store it in a set
ages_set = set()
for age in df['Age']:
    print(age)
    ages_set.add(age)
```
