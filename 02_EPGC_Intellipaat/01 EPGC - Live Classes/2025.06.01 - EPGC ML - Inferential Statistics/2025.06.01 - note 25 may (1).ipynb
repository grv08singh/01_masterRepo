```python
The average salary of a person is hundred dollars back when the survey was conducted 
in 2020 in fresh survey conducted in 2025 with 30 participants, it was found out that 
the average salary of a person is one $40 with a standard deviation of $20 can you 
calculate whether this is a significant difference in salary or not at a 
confidence interval of 95%?
```

```python
null(h0) : (mu = 100) (no significant change in salaries)
alt(h1)  : (mu != 100) (significant changes in salaries)
```

```python
#lib
import numpy as np
from scipy import stats

#data
population_mean = 100 #2019
sample_mean = 87 #2025
n = 30
std_dev = 20
alpha = 0.05

#t_statistic
t_statistic = (sample_mean - population_mean)/(std_dev/np.sqrt(n))
print(f"t_statistic : {t_statistic}")

#degree of freedom
df = n - 1
print(f"df : {df }")

# #t_critical_Value percent point function
# #if 2 tail test - alpha/2.    if one tail
# t_critical_value_upperbound = stats.t.ppf(1 - alpha/2, df)
# t_critical_value_lowerbound = stats.t.ppf(alpha/2, df)
# print(f"t_critical_value_upperbound : {t_critical_value_upperbound}")
# print(f"t_critical_value_lowerbound : {t_critical_value_lowerbound}")


#if you impo 
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic),df)) #syntax
print(f"p_value : {p_value}")

#universal line
if p_value < alpha:
    print("Reject H0 : (mu != 100) (significant changes in salaries)")
else:
    print("Accept H0: (mu = 100) (no significant change in salaries)")
```

```python
A factory has a machine that dispenses 80 ML of a shampoo in a bottle and an employee believes that 
the average amount of shampoo is not ATM but 78ML using the sample size of 40 samples and a standard 
deviation of 2.5 ML. can you state the null and alternate hypothesis at 95% confidence interval and provide 
enough evidence to support the idea that the machine is working properly or is not working properly
```

```python
null : avg shampoo = 80 (machine is working fine)
alt : avg shampoo != 80 (machine is not working fine)
```

```python
import numpy as np
from scipy import stats

#data
sample_mean = 79.2    #0.8 ml  rejet acctp
population_mean = 80
n= 40
alpha = 0.05
std_dev = 2.5

#z_stat
z_stat = (sample_mean - population_mean) / (std_dev/np.sqrt(n))
print(f"z_stat : {z_stat}")

# #do you need critical value? NO
# z_critical_value = stats.norm.ppf(1 - alpha/2)
# print(f"z_critical_value : {z_critical_value}")

p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
print(f"p_value : {p_value}")

if p_value < alpha:
    print("Reject H0 :  avg shampoo != 80 (machine is not working fine)")
else:
    print("Accept H0: avg shampoo = 80 (machine is working fine)")
```

```python
In a sample survey conducted in Bangalore, it was found that 960 people out of 1860 were vegetarian 
and rest were non-vegetarian. Can you find out whether both vegetarian and non-vegetarian are 
equally popular in Bangalore at 1% level of significance #99% CI
```

```python
null :  veg and non-veg are 50-50 in prop 
alt :   veg and non-veg are not in 50-50 prop (they are signisficantly diff)
```

```python
import numpy as np
#new statsmodel
from statsmodels.stats.proportion import proportions_ztest

vegetarian_count = 960
total_pop = 1860
alpha = 0.01

p_hat = vegetarian_count / total_pop
print(f"p_hat : {p_hat}")

# nobs = numbers of observation

z_statistic, p_value = proportions_ztest(count=vegetarian_count,
                                        nobs=total_pop,
                                        value=0.50,  #veg-50 - 50
                                        alternative="two-sided")

print(f"z_statistic : {z_statistic}")
print(f"p_value : {p_value}")

if p_value < alpha:
    print("Reject H0 :veg and non-veg are not in 50-50 prop (they are signisficantly diff) ")
else:
    print("Accept H0:veg and non-veg are 50-50 in prop  ")
```

```python
# ftest

null = var_oroe and var parle are equall (similar)
alternate : var_oero != var_aprle (sig diff from each other)
```

```python
import numpy as np
from scipy import stats

#data
n1 = 10
n2 = 13
x1 = 15
x2 = 14
ssd1= 90
ssd2 = 108
alpha = 0.05

#calc var
var_oreo = ssd1 / (n1 - 1)
var_parle = ssd2 / (n2 - 1)
print(f"var_oreo:{var_oreo} || var_parle:{var_parle}")

f_stats = max(var_oreo, var_parle)/min(var_oreo,var_parle)
print(f"f_stats:{f_stats}")

df_oreo = n1 - 1
df_parle = n2 - 1

p_value = 2 * (1 - stats.f.cdf(f_stats, df_oreo, df_parle))
print(f"p_value:{p_value}")

if p_value < alpha:
    print("Reject H0 :var_oero != var_aprle (sig diff from each other)")
else:
    print("Accept H0: var_oroe and var parle are equall (similar)  ")
```

```python
#Anova - GARDENING

null: fertilizer_A = fertilizer_B = fertilizer_C (all fert are performing the same)
alt : fertilizer_A != fertilizer_B != fertilizer_C
```

```python
from scipy.stats import f_oneway

fertilizer_A = [12, 13, 15, 14, 16, 13, 15, 14, 15, 13]
fertilizer_B = [14, 15, 16, 15, 17, 15, 16, 14, 16, 15]
fertilizer_C = [11, 12, 13, 12, 14, 12, 13, 11 ,13 ,12]

#one way anova
f_statistics, p_value = f_oneway(fertilizer_A,fertilizer_B,fertilizer_C)
print(f"f_statistics : {f_statistics}")
print(f"p_value : {p_value}")

if p_value < 0.05:
    print("Reject H0 :fertilizer_A != fertilizer_B != fertilizer_C")
else:
    print("Accept H0:fertilizer_A = fertilizer_B = fertilizer_C (all fert are performing the same)")
```

```python
import numpy as np
from scipy.stats import chi2
observed = np.array([23, 16, 14, 19, 28])
expected = np.array([42, 32, 27, 37, 39]) #riya 

chi2_stat = np.sum((observed-expected)**2 / expected)
chi2_stat

alpha = 0.05
df = 4
p_value = 1 - chi2.cdf(chi2_stat, df)
print(f"p_value :{p_value}")

if p_value < alpha:
    print("Reject H0 : OB != Ex (Do not purchase)")
else:
    print("Accept H0:OB = Ex (purchase)")
```
