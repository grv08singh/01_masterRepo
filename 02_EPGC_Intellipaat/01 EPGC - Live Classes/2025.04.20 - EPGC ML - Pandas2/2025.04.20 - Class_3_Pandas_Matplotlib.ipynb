# Matplotlib

```python
- It is an open source library 
- It is freely available 
- It is used for 2d plot 
- Few customization are available 

Color code

r - red
g - green
b - blue
k - black
m - managenta
c - cyan 

marker 

o - circle marker 
+ - plus marker 
D - diamond marker 
s - square marker
^ - triangle marker 

line style 

-   -- solid line 
--  --  dash line 
-.  --  dash dot line 
```

# Line plot

```python
# import libraries
import matplotlib.pyplot as plt 
import numpy as np 

# creating data 
a = np.linspace(0,10,20)
# print(a)

# figure size 
# plt.figure(figsize = (6,4))

# axes 
ax1 = plt.subplot()

ax1.plot(a, linewidth = 1, linestyle = '--', color = 'r', marker = 'o', alpha = 0.8)
plt.xlabel('No. of points')
plt.ylabel('Data Values')
plt.title('Line Plot')
plt.grid(True)
plt.savefig('Lineplot.png')
plt.show()

# Draw five lines which are non paraller and non intersecting
```

```python
a = np.linspace(0, 10, 20)
a1 = np.linspace(3, 11, 20)
a2 = np.linspace(2, 10.5, 20)
a3 = np.linspace(1, 10.3, 20)
a4 = np.linspace(10, 5, 5)
plt.figure(figsize=(6, 4))
ax1 = plt.subplot()
ax1.plot(a)
ax1.plot(a1)
ax1.plot(a2)
ax1.plot(a3)
ax1.plot(a4)
plt.xlabel('No of points')
plt.ylabel('Data Values')
plt.title('Line Plot Demo')
plt.grid(True)
plt.show()
```

```python
a = np.linspace(0, 10, 20)
a1 = np.linspace(3, 11, 20)
a2 = np.linspace(2, 10.5, 20)

plt.figure(figsize=(10, 4))

ax1 = plt.subplot(131) # 1- rows, 3 - column 1 - position
ax2 = plt.subplot(132) # 1- rows, 3 - column 2 - position
ax3 = plt.subplot(133) # 1- rows, 3 - column 3 - position

ax1.plot(a)
ax2.plot(a1)
ax3.plot(a2)

plt.show()
```

```python
a = np.linspace(0, 10, 20)
a1 = np.linspace(3, 11, 20)
a2 = np.linspace(2, 10.5, 20)

plt.figure(figsize=(4, 10))

ax1 = plt.subplot(311) # 1- rows, 3 - column 1 - position
ax2 = plt.subplot(312) # 1- rows, 3 - column 2 - position
ax3 = plt.subplot(313) # 1- rows, 3 - column 3 - position

ax1.plot(a)
ax2.plot(a1)
ax3.plot(a2)

plt.show()
```

# Stack plot

```python
x = [1,2,3,4,5]
y = [2,4,5,3,1]

plt.figure(figsize = (6,4))
ax1 = plt.subplot()

ax1.stackplot(x,y, color = 'r', alpha = 0.2)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [[2,4,5,3,1], 
     [2,4,5,3,5],
     [2,2,5,3,1],
     [2,1,5,3,1],
     [1,4,5,2,1]]

plt.figure(figsize = (6,4))
ax1 = plt.subplot()

ax1.stackplot(x,y)
plt.show()
```

```python
# bar chart

x = ['A', 'B', 'C', 'D']
y = [10,20,25,36]

plt.figure(figsize = (6,4))
ax1 = plt.subplot()

ax1.bar(x,y, color = 'r')
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y = [10,20,25,36]

plt.figure(figsize = (6,4))
ax1 = plt.subplot()

ax1.bar(x,y, color = ['r','g'])
plt.show()
```

# boxplot

```python
- it is used to find the outlier
- outlier is the unwanted contaminet which is hampering our result 
```

```python
x = [-125, 10,20,30,21,3,4,24,25,34,40,12,15,18,99,100]

plt.figure(figsize = (6,4))
ax1 = plt.subplot()

ax1.boxplot(x)
plt.show()
```

# Violin Plot

```python
# more the thickness of the violin more will be the data 
# lesser the thickness of violin lesser will be the data and it might be possible 
# that outlier is present

x = [10,20,30,21,3,4,24,25,34,40,12,15,18,99]

plt.figure(figsize = (6,4))
ax1 = plt.subplot()

ax1.violinplot(x)
plt.show()
```
