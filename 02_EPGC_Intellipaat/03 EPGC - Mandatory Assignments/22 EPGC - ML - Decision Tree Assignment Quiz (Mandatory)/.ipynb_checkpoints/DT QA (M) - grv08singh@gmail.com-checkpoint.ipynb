```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
```

```python
df = pd.read_csv('Pokemon.csv')
df.head(2)
```

```python
df.info()
```

```python
df_X = df.drop(columns=['Name','Legendary','Type 2'])
df_y = df[['Legendary']]
```

```python
df_le = LabelEncoder()
df_X['Type 1'] = df_le.fit_transform(df_X['Type 1'])
df_y.head(2)
```

```python
df_ss = StandardScaler()
df_X_scaled = pd.DataFrame(df_ss.fit_transform(df_X))
df_X_scaled.head(2)
```

```python
df_X_scaled_train, df_X_scaled_test, df_y_train, df_y_test = train_test_split(df_X_scaled, df_y, train_size=0.7, random_state=1)
```

```python
dt = DecisionTreeClassifier(random_state=1)
```

```python
dt.fit(df_X_scaled_train, df_y_train)
```

```python
y_pred = dt.predict(df_X_scaled_test)
```

```python
confusion_matrix(df_y_test, y_pred)
```

```python
accuracy_score(df_y_test, y_pred)
```
