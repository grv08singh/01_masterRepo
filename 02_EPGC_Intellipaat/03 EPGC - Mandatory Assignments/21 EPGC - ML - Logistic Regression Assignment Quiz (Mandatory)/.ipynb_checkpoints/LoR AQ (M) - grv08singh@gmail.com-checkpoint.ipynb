```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
```

```python
df = pd.read_csv('census-income .csv')
df.head(2)
```

```python
df.info()
```

```python
df_X = df.drop(columns=['annual_income'])
df_y = df['annual_income']
```

```python
df_le = LabelEncoder()
```

```python
for col in df_X.columns:
    if df_X[col].dtype == 'object':
        df_X[col] = df_le.fit_transform(df_X[col])
```

```python
df_ss = StandardScaler()
```

```python
df_X_scaled = df_ss.fit_transform(df_X)
```

```python
df_X_scaled_train, df_X_scaled_test, df_y_train, df_y_test = train_test_split(df_X_scaled, df_y, test_size=0.3, random_state=0)
```

```python
LoR = LogisticRegression(random_state=0)
```

```python
LoR.fit(df_X_scaled_train, df_y_train)
```

```python
y_pred = LoR.predict(df_X_scaled_test)
pd.DataFrame(y_pred)
```

```python
accuracy_score(df_y_test, y_pred)
```

```python
[tp, fn],[fp, tn] = confusion_matrix(df_y_test, y_pred)
```

```python
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * precision * recall / (precision + recall)

n_prec = tn / (tn + fn)
n_recall_or_specificity = tn / (tn + fp)
```
