```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, accuracy_score
```

## data

```python
data_X = data[['YearsExperience']]
data_y = data['Salary']
data_ss = StandardScaler()
```

```python
scaled_data_X = data_ss.fit_transform(data_X)
```

```python
scaled_data_X_train, scaled_data_X_test, data_y_train, data_y_test = train_test_split(scaled_data_X, data_y, test_size=0.4, random_state=0)
```

```python
data_lr = LinearRegression()
```

```python
data_lr.fit(scaled_data_X_train, data_y_train)
```

```python
data_pred = data_lr.predict(scaled_data_X_test)
```

```python
data_r2 = r2_score(data_y_test, data_pred)
data_r2
```

## housing

```python
housing.head(2)
```

```python
housing.info()
```

```python
housing.columns
```

```python
housing_X = housing.drop(columns=['id','date','price'])
housing_y = housing['price']
```

```python
housing_ss = StandardScaler()
```

```python
scaled_housing_X = housing_ss.fit_transform(housing_X)
```

```python
scaled_housing_X_train, scaled_housing_X_test, housing_y_train, housing_y_test = train_test_split(scaled_housing_X, housing_y, test_size=0.2, random_state=42)
```

```python
housing_lr = LinearRegression()
```

```python
housing_lr.fit(scaled_housing_X_train, housing_y_train)
```

```python
housing_preds = housing_lr.predict(scaled_housing_X_test)
```

```python
r2_score = r2_score(housing_y_test, housing_preds)
r2_score
```
