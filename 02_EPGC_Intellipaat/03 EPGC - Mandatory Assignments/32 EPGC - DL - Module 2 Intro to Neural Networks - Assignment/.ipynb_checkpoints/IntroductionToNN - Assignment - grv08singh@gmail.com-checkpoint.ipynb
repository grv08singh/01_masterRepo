```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Methodology:
#### 1. Activation Functions:
* Define the sigmoid function, a popular activation function used in neural networks.


```python
def sigmoid(z):
    return 1/(1 + math.exp(-z))
```

* Define the tanh function, which is another common activation function that returns values between -1 and 1.


```python
def tanh(z):
    return ((math.exp(z) - math.exp(-z))/(math.exp(z) + math.exp(-z)))
```

* Define the relu function, which returns positive values as they are and zeroes out negative values.


```python
def relu(z):
    return max(0,z)
```

#### 2. Sample Data:
* Use the following sample data for training: x1=3, x2=2, yactual=17

```python
x1 = 3
x2 = 2
y_actual = 17

print(f"x1 = {x1}\nx2 = {x2}\ny_actual = {y_actual}")
```

#### 3. Feed Forward:
* Initialize the random weights w1 and w2 between 1 and 10.

```python
w1 = np.random.randint(1,10,1)[0]
w2 = np.random.randint(1,10,1)[0]
print(f"w1 = {w1}\nw2 = {w2}")
```

* Predict the output ypred using the equation ypred= x1 x w1 + x2 x w2

```python
y_pred = x1*w1 + x2*w2
print(y_pred)
```

#### 4. Error Calculation:
* Compute the squared error as error = (yactual-ypred)*(yactual-ypred)

```python
error = (y_actual - y_pred)**2
print(f"error = {error}")
```

#### 5. Back Propagation:
* Compute the gradients of the error with respect to the weights.

```python
grad_w1 = 2*(y_actual - y_pred)*(-x1)
grad_w2 = 2*(y_actual - y_pred)*(-x2)
print(f"grad_w1 = {grad_w1}\ngrad_w2 = {grad_w2}")
```

* Update the weights using the computed gradients and a learning rate.

```python
learning_rate = 0.01
w1 -= learning_rate*grad_w1
w2 -= learning_rate*grad_w2
print(f"Updated weights: \nw1 = {w1}\nw2 = {w2}")
```

#### 6. Training:
* The above steps (Feed Forward to Back Propagation) are performed iteratively
  (20 times in this example) to refine the weights and minimize the error.

```python
y_pred_history = [y_pred]
error_history = [error]
for epoch in range(20):
    print(f"\nepoch {epoch+1}")
    y_pred = x1*w1 + x2*w2
    y_pred_history.append(y_pred)
    print(f"y_pred = {y_pred}")
    error = (y_actual - y_pred)**2
    error_history.append(error)
    print(f"error = {error}")
    grad_w1 = 2*(y_actual - y_pred)*(-x1)
    grad_w2 = 2*(y_actual - y_pred)*(-x2)
    print(f"grad_w1 = {grad_w1}\ngrad_w2 = {grad_w2}")
    w1 -= learning_rate*grad_w1
    w2 -= learning_rate*grad_w2
    print(f"Updated weights: \nw1 = {w1}\nw2 = {w2}")
```

#### 7. Visualization:
* Plot the progression of the error across epochs.

```python
fig,ax = plt.subplots(figsize=(12,4))
ax.plot(np.arange(21),error_history,marker=".",markersize=10, label="Sum Squared Error", color='red')
plt.xlabel("Epoch Number")
plt.ylabel("Error")
plt.title("Cost vs. Epoch")
plt.legend()
plt.tight_layout()
plt.show()
```

* Plot the progression of ypred across epochs.

```python
fig,ax = plt.subplots(figsize=(12,4))
ax.plot(np.arange(21),y_pred_history,marker=".",markersize=10, label="y_pred", color='blue')
plt.xlabel("Epoch Number")
plt.ylabel("Prediction")
plt.title("y_pred vs. Epoch")
plt.legend()
plt.tight_layout()
plt.show()
```
