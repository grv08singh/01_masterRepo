```python
# Q19.Write a function which finds all pythagorean triplets of triangles 
# whose sides are no greater than a natural number n.

def pythagorean_triplets(n):
    lst = []
    for a in range(3, n+1):
        if a%2==1: #odd
            #pass
            mid = a**2 / 2
            b = mid - 0.5
            c = mid + 0.5
        else: #even
            #pass
            mid = (a/2)**2
            b = mid-1
            c = mid+1
            
        if a<n and b<n and c<n:
            lst.append([a,int(b),int(c)])
    return lst

pythagorean_triplets(1000)
```
